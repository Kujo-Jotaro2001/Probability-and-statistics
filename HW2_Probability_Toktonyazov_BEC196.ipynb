## 1. Основные характеристики выборки

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as ss
from openpyxl import load_workbook
```

```python
# Зададим начальное условие для генератора случайных чисел и создадим выборку объемом n=100 из равномерного на (0,1) распределения: 
np.random.seed(0)
uniform_distribution = np.random.uniform(0, 1, 100)
uniform_distribution
```

```python
# Зададим начальное условие для генератора случайных чисел и создадим выборку объемом n=100 из нормального распределения с параметрами (10, 1):
np.random.seed(0)
normal_distribution = np.random.normal(10, 1, 100)
normal_distribution
```

```python
# Зададим начальное условие для генератора случайных чисел и создадим выборку объемом n=100 из показательного распределения с параметром lamda = 1/10: 
# Первый аргумент функции np.random.exponential задается через 1/lambda, поэтому в аргумент подается значение 10;
np.random.seed(0)
exponential_distribution = np.random.exponential(10, 100)
exponential_distribution
```

```python
# Создадим датафрейм из трех распределений:
df = pd.DataFrame({"uniform distribution": uniform_distribution, "normal distribution": normal_distribution,
                   "exponential distribution": exponential_distribution})
df
```

```python
# Выведем основные характеристики распределений:
df.describe()
```

```python
# Построим диаграмму рассеяния для равномерного распределения:
sns.scatterplot(data=df, x="uniform distribution", y=df["uniform distribution"].index)
plt.title("Диаграмма рассеяния равномерного распределения")
plt.xlabel("Значения величин из выборки")
plt.ylabel("Индексы величин из выборки")
```

```python
# Построим диаграмму рассеяния для нормального распределения:
sns.scatterplot(data=df, x="normal distribution", y=df["normal distribution"].index)
plt.title("Диаграмма рассеяния нормального распределения")
plt.xlabel("Значения величин из выборки")
plt.ylabel("Индексы величин из выборки")
```

```python
# Построим диаграмму рассеяния для показательного распределения:
sns.scatterplot(data=df, x="exponential distribution", y=df["exponential distribution"].index)
plt.title("Диаграмма рассеяния показательного распределения")
plt.xlabel("Значения величин из выборки")
plt.ylabel("Индексы величин из выборки")
```

```python
# Построим гистограмму для равномерного распределения:
first_edge, last_edge = df["uniform distribution"].min(), df["uniform distribution"].max()
n_equal_bins = 10
bin_edges = np.linspace(start=first_edge, stop=last_edge, num=n_equal_bins + 1, endpoint=True)
plt.hist(df["uniform distribution"], bins=bin_edges, rwidth=0.9)
plt.grid(axis='y', alpha=0.75)
plt.title("Гистограмма равномерного распределения")
plt.xlabel('Значения величин из выборки')
plt.ylabel('Частота')
```

## 2. Центральная предельная теорема

```python
# Зададим начальное условие для генератора случайных чисел и создадим матрицу 20х100 с элементами из равномерного на (0, 1) распределения;
# Создадим из матрицы датафрейм для удобства дальнейшего использования;
np.random.seed(0)
uniform_matrix = np.random.uniform(0, 1, (20, 100))
df_uniform = pd.DataFrame(data=uniform_matrix)
df_uniform
```

```python
# Посчитаем среднее значение для каждой из ста выборок:
columns_mean = df_uniform.describe().loc["mean"]
# Посчитаем среднее значение и стандартное отклонение для элементов в columns_mean:
mean_value, std_value = columns_mean.mean(), columns_mean.std()
# Посчитаем стандартизированное выборочное среднее:
standardized_sample_mean = (columns_mean - mean_value) / std_value
standardized_sample_mean
```

```python
# Построим гистограмму стандартизированного выборочного среднего: 
first_edge, last_edge = standardized_sample_mean.min(), standardized_sample_mean.max()
n_equal_bins = 10
bin_edges = np.linspace(start=first_edge, stop=last_edge, num=n_equal_bins + 1, endpoint=True)
plt.hist(standardized_sample_mean, bins=bin_edges, rwidth=0.9)
plt.grid(axis='y', alpha=0.75)
plt.title("Гистограмма стандартизированного выборочного среднего")
plt.xlabel('Значения величин')
plt.ylabel('Частота')
```
